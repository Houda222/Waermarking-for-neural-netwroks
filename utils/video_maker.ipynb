```python
import nibabel as nib
import numpy as np
import cv2
import matplotlib.pyplot as plt
img=nib.load("C:/Users/arifa/OneDrive/Bureau/BRATS_modification/new_data/embedded_data_segs.nii")
```

```python
data=img.get_fdata()
data_array=np.array(data)
plt.imshow(data_array[:,:,70])
plt.show()
```

```python
# Determine the dimensions of the images
def video_maker(name,data_array):
    height, width,time = data_array.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = 24  # Frames per second
    out = cv2.VideoWriter('videos/'+name+".mp4", fourcc, fps, (width, height),isColor=False)
# Write each image to the video file
    for t in range(time):
        wiw=(data_array[:,:,t]*255/4).astype(np.uint8)
        out.write(wiw)
    out.release()
    print("Video created successfully!")
```

```python
video_maker("seg_video",data_array)
```
